```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 11)
y = x**2
```

```python
k = np.linspace(-4*np.pi, 4*np.pi, num=100)
z = np.sin(1.5*k)
m = np.tan(k)
```

#### Subplot
Creates stacked graphs, adding up rows or columns, each with different data.

```python
plt.subplot(1,2,1)
plt.hist(y, color='#803763')
plt.subplot(1,2,2)
plt.plot(k, z, 'g--')
plt.plot(k, m, 'b-.')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.hist(y, color='black')
plt.subplot(2,1,2)
plt.plot(k, z, 'g--')
plt.plot(k, m, 'b-.')  
plt.show()
```

```python
# Generate some data
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and add subplots
fig = plt.figure()

# Add the first subplot
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(x, y1)
ax1.set_title('Sine')

# Add the second subplot
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(x, y2)
ax2.set_title('Cosine')

# Adjust the layout
plt.tight_layout()

# Display the figure
plt.show()
```
