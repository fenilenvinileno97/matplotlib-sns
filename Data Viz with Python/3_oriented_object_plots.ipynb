```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

There are different ways to plot in Matplotlib, we've already checked up pyplot; nevertheless, customize and accessing to some plot parameters can be quite complex.

#### Oriented object Matplotlib method
An object defines a figure, a fig is an empty canvas which we can insert different *axes*, each one owns their own *axis*.

Figure represents everything, we'll set up plot within it. We can make a much better tailoring in each subplot using oriented object methods.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes2 = fig.add_axes([0.15,0.8,0.3,0.15])
axes.plot(x,y, 'b')
axes2.plot(y,x, 'r:')
axes2.set_facecolor('grey')
fig.show()
```

#### Plotting acceleration and velocity respect to time

In kinematics, we know that change of distance versus time is defined as velocity, $\frac{dx}{dt} = V$, which integrated from $t_0$ to $t_f$ yields to:

$x = v(t-t_0) + x_0$

$\frac{dx}{dt} = \frac{dv}{dt}(t-t_0)$

$\frac{dv}{dt}$ is defined as acceleration $a$, so, integrating an replacing $x_0 = v_0t$, yields:

$\Delta x = v_0 t + \frac{1}{2} a t^2$


```python
time = np.arange(0., 10., 0.2)
velocity = np.zeros_like(time, dtype=float)
distance = np.zeros_like(time, dtype=float)
```

```python
g = 9.8
velocity = g * time
distance = 0.5 * g * np.power(time, 2)
```

```python
plt.figure(figsize=(6,5), dpi=100)
plt.plot(time,distance,'g-.')
plt.xlabel("Time (s)")
plt.ylabel("Distance (m)")
plt.legend(["Distance"])
plt.grid(True)
plt.show()
```

The plot shows how much distance was covered by the free-falling object with each passing second.

```python
plt.figure(figsize=(6,5), dpi=100)
plt.plot(time, velocity,'r--')
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.legend(["Velocity"])
plt.grid(True)
```

```python
plt.figure(figsize=(6,5), dpi=100)
plt.plot(time, velocity,'g-')
plt.plot(time, distance,'b-')
plt.ylabel("Distance and Velocity")
plt.xlabel("Time")
plt.legend(["Distance", "Velocity"])
plt.grid(True)
```

```python
fig, ax1 = plt.subplots()

ax1.set_ylabel("distance (m)", color="blue")
ax1.set_xlabel("time")
ax1.plot(time, distance, "blue")
ax1.set_yticks(np.linspace(*ax1.get_ybound(), 10))
ax1.tick_params(axis="y", labelcolor="blue")
ax1.xaxis.grid()
ax1.yaxis.grid()

ax2 = ax1.twinx() # create another y-axis sharing a common x-axis


ax2.set_ylabel("velocity (m/s)", color="green")
ax2.set_xlabel("time")

ax2.tick_params(axis="y", labelcolor="green")
ax2.plot(time, velocity, "green")
ax2.set_yticks(np.linspace(*ax2.get_ybound(), 10))

fig.set_size_inches(7,5)
fig.set_dpi(100)
fig.legend(["Distance", "Velocity"])
plt.show()
```
