```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### How to contextualize our plots
In order to provide context to our graphs, we need to use *titles*, *legends*, *size*, or *labels*, so that our graph will have a broader context.

```python
pristine_mcc = pd.read_csv('./pristine_mcc.csv', sep=',', header=0)
jolyg_mcc = pd.read_csv('./jolyg_mcc.csv', sep=',', header=0)
```

We create our FT-IR subplots based on provided datasets.

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(pristine_mcc['Wavelength'], pristine_mcc['Transmittance'], 'b')
ax2.plot(jolyg_mcc['Wavenumber'], jolyg_mcc['Transmittance'], 'r')
fig.tight_layout()
fig.show()
```

We can create titles using `.set_title()` and also set limits in axis using `.set_xlim()`. In FT-IR plots, x usually ranges from 4000 to 400 units.

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(pristine_mcc['Wavelength'], pristine_mcc['Transmittance'], 'b')
ax2.plot(jolyg_mcc['Wavenumber'], jolyg_mcc['Transmittance'], 'r')
ax1.set_xlim(4000,400)
ax1.set_title('Pristine MCC')
ax2.set_xlim(4000,400)
ax2.set_title('Lauric esterified MCC')
fig.tight_layout()
fig.show()
```

Using `.set_xlabel()` and `.set_ylabel()` methods, we can match each axis with a description.

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(pristine_mcc['Wavelength'], pristine_mcc['Transmittance'], 'b')
ax2.plot(jolyg_mcc['Wavenumber'], jolyg_mcc['Transmittance'], 'r')
ax1.set_xlim(4000,400)
ax1.set_title('Pristine MCC')
ax1.set_xlabel('Wavenumber (cm-1)')
ax1.set_ylabel('Transmittance (%)')
ax2.set_xlim(4000,400)
ax2.set_title('Lauric esterified MCC')
ax2.set_xlabel('Wavenumber (cm-1)')
ax2.set_ylabel('Transmittance (%)')
fig.tight_layout()
fig.show()
```

Also, we can change figure size, indicating `figsize = (width, height)`within `.subplots()` or can be included in `.figure(figsize)` method.

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize= (1,7))
```

```python
fig = plt.figure(figsize=(1,7))
fig.subplots(1,2)
fig.show()
```

Including legends can help to specify plots nature when we plot several graphs in a same axe.

```python
plt.figure(figsize=(10,4), dpi=300)
plt.plot(pristine_mcc['Wavelength'], pristine_mcc['Transmittance'], 'b', label='Pristine MCC')
plt.plot(jolyg_mcc['Wavenumber'], jolyg_mcc['Transmittance'], 'r', label='Lauric MCC')
plt.xlim(4000,400)
plt.xlabel('Wavenumber (cm$^{-1}$)')
plt.ylabel('Transmittance')
plt.legend()
plt.tight_layout()
plt.show()
```
