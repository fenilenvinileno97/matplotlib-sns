```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

##### Scatter plot

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
sns.lmplot(data=tips, x='total_bill', y='tip', hue='day')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

```python
markers = {"Lunch": 'D', "Dinner": 's'} #we have to check up markers format in documentation
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size', markers=markers)
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

##### Line plot

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', size='size', style='time')
plt.legend(loc= 'center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

##### Relplot

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='time', kind='line')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='time', kind='scatter', col='time')
plt.show()
```
