````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

#### Seaborn

Seaborn is a library for making statistical graphics in ```Python```. It's bulit on top of ```matplotlib``` and integrates closely with ```pandas``` data structures.

Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API let focus on what the different elements of our plots mean, rather than on the details of how to draw them.

##### Advantages of ```Seaborn```

* It's fast
* Code inputs are quite easy and not so specific as `matplotlib`
* Highly customizable graphs and visualizations.

#### Plots in `Seaborn`

Seaborn offers some main features for specific data visualization.

* Relational plots or diagrams
* Data distributions
* Categoric variables

For instance:

* **Relplot (relational)**: scatterplot, lineplot
* **Displot (distribution)**: histplot, kdeplot, ecdfplot, rugplot.
* **Catplot (categorical)**: stripplot, swamplot, boxplot, violinplot, pointplot, barplot.

````python
food_prod = pd.read_csv('../Datasets/Food_Production.csv', sep=',', header=0, encoding='utf-8')
reduced_table = food_prod[['Food product', 'Land use change', 'Total_emissions']].sort_values(by='Total_emissions', ascending=False).head(5)
reduced_table
````

````python
sns.set_theme(style='dark', palette='flare', font='monospace', font_scale=1)
sns.barplot(x=reduced_table['Food product'], y=reduced_table['Total_emissions'])
plt.tight_layout()
plt.show()
````
