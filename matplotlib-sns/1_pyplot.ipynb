```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(start=0, stop=5, num=11)
y = x**2
```

```python
plt.plot(x,y, 'ro')
plt.show
```

```python
plt.hist(y)
plt.show
```

```python
plt.pie(y)
plt.show
```

```python
plt.boxplot(x)
plt.show
```

```python
plt.scatter(x,y)
plt.show
```

```python
x.mean()
```

```python
k = np.linspace(-4*np.pi, 4*np.pi, num=100)
z = np.sin(1.5*k)
```

```python
fig, ax = plt.subplots()
ax.plot(k,z, linewidth=2.0, color='#803763')
# ax.set(xlim=(0,1), xticks=np.arange(-10,10),
#        ylim=(0,1), yticks=np.arange(-2,3))
plt.show()
```
