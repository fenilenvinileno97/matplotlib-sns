```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.info()
```

##### Count plot

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

##### Swarm plot

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
plt.legend()
plt.show()
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.legend()
plt.show()
```

##### Box plot

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', showfliers=True)
plt.legend()
plt.show()
```

We can include swarmplots for infering more info.

```python
plt.figure(dpi=200)
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', showfliers=True)
sns.swarmplot(data=tips,x='day',y='total_bill', hue='sex',dodge=True, palette="dark:black",alpha=0.7)
plt.legend()
plt.show()
```

##### Violin plot

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=False)
plt.legend()
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.legend()
plt.show()
```

##### Catplot

```python
sns.catplot(data= tips, x= 'day', y = 'total_bill', hue= 'sex', dodge=True, kind= 'box', col= 'time')
plt.show()
```
