```python
import seaborn as sns
import matplotlib.pyplot as plt
```

`'tips'` is a default preloaded dataset in seaborn library (requires internet).

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

This plot stands for a total bill hystogram, when it is shown the distribution of maximum tips given by all users (male and female) in the restaurant (independent of meal time), and also the lower and upper tails.

```python
sns.displot(data= tips, x= 'total_bill')
plt.show()
```

From below plot, a bivariated distribution, we can observe a relationship of tips per person and totalized bills independent of smoker users and meal time.

```python
sns.displot(data= tips, x= 'total_bill', y= 'tip');
```

In this case, a bivariated distribution of tips/person and total bills will be categorized in female and male. As we can observe in the plots below, including density distributions, female and male users behavior **tends to be similar in terms of tips payment**, but we can appreciate that male users are used to pay more than female users.

```python
sns.displot(data= tips, x= 'total_bill', y= 'tip', hue = 'sex')
plt.show()
```

```python
sns.displot(data= tips, x= 'total_bill', hue = 'sex', kind = 'kde', fill=True)
plt.show()
```

```python
sns.displot(data= tips, x= 'total_bill', hue = 'sex', kind = 'kde', legend=True, palette='dark', alpha = 0.5, fill=False)
plt.show()
```
