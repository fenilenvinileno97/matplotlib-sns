```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1.5, 1.5, num=500)
y = (x**2 * (1 - x**2)**3)
```

```python
for k in enumerate(plt.style.available):
    print(k)
```

```python
for i in plt.style.available[3:20]:
    fig, ax = plt.subplots(figsize=(2,2), dpi=100)
    ax.plot(x,y)
    ax.plot(y,x)
    ax.set_title(i, color='purple')
    fig.tight_layout()
    plt.style.use(i)
plt.close()
```

```python
k = np.linspace(-10, 10, num=10)
```

```python
fig, ax = plt.subplots(figsize = (5,5), dpi=100)
ax.plot(k,k+1, 'r--')
ax.plot(k,k+2, 'bo-')
ax.plot(k,k+3, 'g.:')
ax.plot(k,k+4, 'purple')
fig.show()
```

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.plot(k,k+3,color = '#D426C8') 
ax.plot(k,k+6,color = '#66FF89')
ax.plot(k,k+9,color = 'blue')
ax.plot(k,k+12, color = 'black')
fig.show()
```

```python
fig, ax = plt.subplots(figsize = (4,4))
ax.plot(k,k+3,color = 'blue', linestyle = 'dashed', marker = 'x')
ax.plot(k,k+12, color = 'black',linestyle = '--', marker = 'P')
fig.show()
```

```python
fig, ax = plt.subplots(figsize = (4,4))
ax.plot(k,k+3,color = 'blue', linestyle = 'dashed', marker = '8', markersize = 10, markerfacecolor= "#37D842")
ax.plot(k,k+12, color = 'black',linestyle = '--', marker = 'P', markerfacecolor="#FF0000")
fig.show()
```
