```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

#### Histogram

```python
plt.plot(figsize=(3,3))
plt.hist(data, bins=20)
plt.show()
```

#### Step plot

```python
plt.plot(figsize=(3,3))
plt.hist(data, bins=20, histtype='step')
plt.show()
```

#### Boxplot

```python
plt.plot(figsize=(3,3))
plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
modified_data = np.append(data, 120)
plt.boxplot(modified_data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

#### Scatter plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area)
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'v')
plt.show()
```

```python
plt.scatter(x,y, s=area, c= colors, marker = 'o', alpha= 0.5)
plt.show()
```
