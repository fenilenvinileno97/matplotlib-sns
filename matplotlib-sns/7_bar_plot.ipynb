```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
food_prod = pd.read_csv('../Datasets/Food_Production.csv', sep=',', header=0, encoding='utf-8')
reduced_table = food_prod[['Food product', 'Land use change', 'Total_emissions']].sort_values(by='Total_emissions', ascending=False).head(5)
reduced_table
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(7,5), dpi=200)
ax1.bar(reduced_table['Food product'], reduced_table['Total_emissions'], width=0.5, color=['aqua', 'grey', 'teal', 'crimson', 'violet'])
ax1.set_ylabel('Total CO₂ emissions (tons)')
ax1.set_xlabel('Food product')
ax2.bar(reduced_table['Food product'], reduced_table['Land use change'], width=0.5, color='red')
ax2.set_ylabel('Land use change (m2)')
ax2.set_xlabel('Food product')
fig.tight_layout()
```

```python
land_use = pd.read_csv('../Datasets/Land.csv', sep=',', header=0, encoding='utf-8')
land_use['Series'].value_counts()
```

```python
arable_col = land_use[(land_use['Land'] == 'Colombia') & (land_use['Series'] == 'Arable land (% of total land area)')]
arable_col
```

```python
biodiversity_col = land_use[(land_use['Land'] == 'Colombia') & (land_use['Series'] == 'Important sites for terrestrial biodiversity protected (% of total sites protected)')]
biodiversity_col
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(7,5), dpi=200)
ax1.bar(arable_col['Year'], arable_col['Value'], width=0.5, color=['#097969', '#50C878', '#4CBB17'])
ax1.set_title('Land use in Colombia')
ax1.set_xlabel('Years')
ax1.set_ylabel('Arable land (% of total land)')
ax2.bar(biodiversity_col['Year'], biodiversity_col['Value'], width=0.5, color=['green'])
ax2.set_title('Important sites for terrestrial biodiversity protected')
ax2.set_xlabel('Years')
ax2.set_ylabel('% of total protected sites')
fig.tight_layout()
```
