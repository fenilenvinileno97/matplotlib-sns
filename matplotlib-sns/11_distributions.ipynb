````python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
````

````python
tips = sns.load_dataset('tips')
tips['tip'].min()
````

````python
tips['tip'].max()
````

#### `.histplot()`
As we saw above, minimum tip/person value is 1.0 USD and maximum tip value is 10.0 USD, so when we use a cumulative histogram, we can appreciate that all poblations are summed up (relative frequencies) until reaching total frequency.

````python
sns.histplot(data=tips, x='tip', bins=15, cumulative=True)
plt.show()
````

````python
sns.histplot(data=tips, x='tip', bins=20, hue='sex', stat='density', multiple='dodge', cumulative=False)
plt.show()
````

#### Density plot `.kde()`

````python
sns.kdeplot(data=tips, x='tip', hue='sex')
plt.show()
````

With cumulation (installing of scipy in the notebook is key, using ```%pip install scipy``` and then upgrading ```%pip install --upgrade seaborn scipy```)

````python
plt.figure(figsize=(4,3), dpi=100)
sns.kdeplot(data=tips, x='tip', hue='sex', legend=True, cumulative=True)
plt.show()
````

````python
sns.kdeplot(data=tips, x='tip', hue='sex', fill=True)
plt.show()
````

````python
sns.displot(data = tips, x = 'tip', hue = 'sex',kind= 'hist', multiple= 'stack')
plt.show()
````
